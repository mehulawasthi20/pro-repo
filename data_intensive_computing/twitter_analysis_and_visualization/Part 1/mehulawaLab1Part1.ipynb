# Problem1

Defines sales1 using vector initialization, sales2 is initialised using the rpois function which generates 12 numbers (first parameter) with the given mean (2nd parameter). The par function is used to set graphical parameters with bg = cornsilk  which sets the background of the device region to corn silk.  Then sales1 is plotted on the line graph, and sales2 is plotted on the same axes. The type option ’o’ allows for  both points and lines.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")

plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

The following code uses box plot to compare the two sales data. It then uses the text function to add text to the graph

```r
fn<-boxplot(sales,col=c("orange","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

# Problem 4

Following code is used to demonstrate the use of read_csv function. It can be used to read data from csv files and automatically convert into tabular format.Then data is plotted on the graph using the functions plot(to plot the apple data) and lines function to plot a line graph on the axes

# Problem 6

The following code first creates a list of cities using vector initiliazation. Then the ggmap library - geocode is used for getting the latitude and longitude of these cities, for this, a Google developers account is required, then the map function is used to plot the world map, (first parameter of map function). The lat/long points gathered from the geocode function are plotted on this map 





```r
library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "New York", "Buffalo", "Dallas, TX")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit.x,visit.y, col="yellow", pch=36)
```
